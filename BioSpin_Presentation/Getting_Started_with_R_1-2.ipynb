- 백터 
- 백터란 특수한 형태로 설문조사의 선택문항등 카테코리를 저장할 수 있는 벡터인 factor라는 형태가 있다
- factor(대상, 레벨) 을 사용하면 된다
- levels(대상) <- c(바꾸고 싶은 레벨) 을 입력하면 된다


```r
pain <- c(0,3,2,2,1)
fpain <- factor(pain,levels=0,3)
levels(fpain) <- c("none","mild","medium","severe")
fpain
```

```r
as.numeric(fpain)
```

```r
levels(fpain)
```
